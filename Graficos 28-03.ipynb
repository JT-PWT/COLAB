```python
#definir a largura da barra
bin_width = 5

#definir onde começa a primeira barra e onde termmina a última barra
start_bin = 0
end_bin = 10
min_val = min(df['nota'])
max_val = max(df['nota'])


bins = []
atual = start_bin
while atual <= end_bin:
  bins.append(atual)
  atual += bin_width
plt.hist(df['nota'], edgecolor='black', linewidth=1.2, bins = bins)
plt.show()
```

```python
#Plotar as diferenças entre salários por estados
salario_por_estado = df.groupby('estado')['salario'].mean().reset_index()

plt.figure(figsize=(10, 6)) #ajustar o tamanho da figura
plt.bar(salario_por_estado['estado'], salario_por_estado['salario'])
plt.title('Diferenças de salários por estados')
plt.xlabel('Estados')
plt.ylabel('Salário médio')
plt.xticks(rotation=90) # muda rótulos do eixo x
plt.tight_layout() # ajustar o layout para evitar que os rótulos fiquem cortados
plt.show()
```

```python
#salário de produção de SP e MG
df_produção = df[df['departamento']=='Produção']


#agrupar
salarios_por_estado = df_produção.groupby('estado')['salario'].mean().reset_index()

#filtrar apenas SP e MG
df_sp_mg = salarios_por_estado[salarios_por_estado['estado'].isin(['SP', 'MG'])]

plt.figure(figsize=(8, 6)) #ajustar o tamanho da figura
plt.bar(df_sp_mg['estado'], df_sp_mg['salario'])
plt.title('Salários médios do departamento de produção em SP e MG')
plt.xlabel('Estados')
plt.ylabel('Salário médio')
plt.xticks(rotation=90) # muda rótulos do eixo x
plt.tight_layout() # ajustar o layout para evitar que os rótulos fiquem cortados
plt.show()
```
